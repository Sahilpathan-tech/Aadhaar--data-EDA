```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("C:/DSA/aadhar analysis.csv")

# Basic Overview
print("Dataset Shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nMissing values:\n", df.isnull().sum())
print("\nData types:\n", df.dtypes)
print("\nDescriptive Statistics:\n", df.describe(include='all'))

# Convert 'Pin Code' to numeric if possible
df['Pin Code'] = pd.to_numeric(df['Pin Code'], errors='coerce')


```

```python
# Distribution of Age
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Gender Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
# Aadhaar Generation vs Rejection
plt.figure(figsize=(10, 5))
sns.histplot(df[['Aadhaar generated', 'Enrolment Rejected']], bins=30, kde=False)
plt.title("Aadhaar Generated vs Rejected")
plt.legend(["Generated", "Rejected"])
plt.show()
```

```python
# State-wise Aadhaar generation
statewise = df.groupby('State')['Aadhaar generated'].sum().sort_values(ascending=False).head(10)
statewise.plot(kind='barh', figsize=(10, 6), color='skyblue')
plt.title("Top 10 States by Aadhaar Generated")
plt.xlabel("Count")
plt.show()
```

```python
# Correlation heatmap for numeric features
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
